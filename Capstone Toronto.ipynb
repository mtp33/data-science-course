## Exercise 1

Import required libraries

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

Extract hmtl code of the relevant website for scraping

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
data  = requests.get(url).text 
```

Create a BS object from the html code

```python
soup = BeautifulSoup(data,"html5lib")
```

Find all tables on the website and show the 1st one

```python
tables = soup.find_all('table')
print(tables[0].prettify())
```

Scrape relevant data from the 1st table

```python
# List of neighborhoods
table_contents=[]

for row in tables[0].findAll('td'):
    # Empty dictionary for the current neighborhood
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        # Extract relevant text and store into the dictionary
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
```

```python
table_contents
```

Convert to a dataframe

```python
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
pd.set_option('display.max_rows', None)
df.sort_values(by="PostalCode", inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

Impport the longitude and lattidude data for the postcodes from a csv

Join with the neighborhoods dataframe on post code

```python
neighborhoods = pd.concat([df, df2], axis=1, join="inner")
neighborhoods = neighborhoods.drop('Postal Code', axis=1)
neighborhoods
```

```python
from geopy.geocoders import Nominatim
import folium
```

```python
address = 'Toronto'

geolocator = Nominatim(user_agent="explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
```

```python
map1 = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, postalcode in zip(neighborhoods['Latitude'], neighborhoods['Longitude'], neighborhoods['Borough'], neighborhoods['PostalCode']):
    label = '{}, {}'.format(postalcode, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map1)  
    
map1
```
